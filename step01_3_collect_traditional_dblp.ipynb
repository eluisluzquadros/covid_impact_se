### Import das dependencias

```python
import os

from dataprep.connector import connect
from dataprep.eda import plot

conn = connect("dblp", _concurrency = 6)

from google.colab import drive
```

### Collect dblp

```python
df1 = await conn.query("publication", q="covid", _count=20000)
print(df1.shape);df1.head(3)
```
